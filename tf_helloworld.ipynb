```python
# Importa la libreria TensorFlow, muestra la version instalada
# y verifica si hay GPU disponible

import numpy as np
import tensorflow as tf
print(tf.config.list_physical_devices('GPU'))
print(tf.__version__)
```

```python
from sionna.phy.fec.polar.encoding import PolarEncoder
from sionna.phy.fec.crc import CRCEncoder

N = 1024 # Longitud de bloque
K = 512 # Rate 1/2
CRC_tipo = "CRC24A" # CRC de 24 bits, seccion 5.1 3GPP TS 38.212
K_datos = 488 # K - CRC
```

```python
# Generador 'dummy' de bits congelados

import csv

digits = np.random.permutation(N)[:(N-K)]
with open('random_digits.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(digits)
```

```python
CRC_codif = CRCEncoder(CRC_tipo)
bits_congelados = np.loadtxt('random_digits.csv', delimiter=',', dtype=int)
Polar_codif = PolarEncoder(bits_congelados, N)

datos = tf.random.uniform([1, K_datos], minval=0, maxval=2, dtype=tf.int32)
datos = tf.cast(datos, tf.float32)  # convert to float tensor (0.0 or 1.0)

data_con_crc = CRC_codif(datos)      # output shape: [batch_size, K_total]
bits_codif = Polar_codif(data_con_crc)   # output shape: [batch_size, N]

print(datos.shape)  # should be (1, 488) for this example
print(bits_codif.shape)  # should be (1, 1024) for this example
bits_codif
```
